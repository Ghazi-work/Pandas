### Handling missing values


##### we can handle missing values by:

1. Dropping the column or those rows
2. Filling it with mean, median or mode


```python
titanic_df.isnull().sum() / len(titanic_df) * 100
```

dropping dec column cuz too many missing values


```python
titanic_df.drop('deck', axis=1, inplace=True)
```

```python
titanic_df.columns
```

```python
titanic_df.isnull().sum()
```

now we will target age column


```python
print(titanic_df['age'].mean())
print(titanic_df['age'].median())
print(titanic_df['age'].mode()[0])
```

We can fill the age missing values by either **Mean** **Median** **Mode**


```python
titanic_df['age'].fillna(titanic_df['age'].mean(), inplace=True)
```

```python
titanic_df.isnull().sum()
```

```python
titanic_df['embarked'].unique()
titanic_df['embark_town'].unique()
```

I think I see a relation between these two columns its the first letter of the value

I would like to look the rows which are null


```python
selected_columns = titanic_df[['embarked', 'embark_town']]

# Create a boolean mask for rows with missing values in the selected columns
missing_values_mask = selected_columns.isnull().any(axis=1)

# Use the mask to filter the DataFrame and display only rows with missing values
rows_with_missing_values = selected_columns[missing_values_mask]

# Display the result
print(rows_with_missing_values)
```

```python
titanic_df['embarked'].fillna(titanic_df['embarked'].mode()[0], inplace=True)
titanic_df['embark_town'].fillna(
    titanic_df['embark_town'].mode()[0], inplace=True)
```

```python
titanic_df.isnull().sum()
```

Creating bins and labels for histogram (Classes for numerical data)


```python
# 0-5 5-13 13-18 18-25 25-30 30-40 40-50 50+
bins = [0, 5, 13, 18, 25, 30, 40, 50, 60]
labels = ['infant', 'pre-teen', 'teen', 'adult',
          'mature', 'matureII', 'old', 'senior']
titanic_df['Age_classes'] = pd.cut(titanic_df['age'], bins=bins, labels=labels)
titanic_df.head()
```

```python
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.histplot(titanic_df['Age_classes'])
plt.title('Histogram of age groups')
plt.xlabel('Age classes')
plt.ylabel('Count')
plt.show()
```

```python
titanic_df['Age_classes'].value_counts().sort_values()
```

```python
titanic_df.head()
```

Need to find out which how many infants survived or not survived


```python
titanic_df.groupby(['Age_classes', 'survived']).count()
```

```python
# Assuming 'Age_classes' and 'survived' are columns in your DataFrame
result = titanic_df.pivot_table(
    index='Age_classes', columns='survived', aggfunc='size', fill_value=0)

# Display the result
print(result)
```

```python
# 1 represents survived in this example
infant_survived_count = result.loc['infant', 1]
print(f"Number of infants who survived: {infant_survived_count}")
```
