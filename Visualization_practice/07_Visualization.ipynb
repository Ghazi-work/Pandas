```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as ex
import plotly.graph_objects as go
```

```python
df = sns.load_dataset('titanic')
```

```python
df.info()
```

#### We can give hue to the plot based on categorical variable

```python
sns.scatterplot(data=df, x='age', y='fare' ,hue='sex')
```

#### Size of the plot can be changed using plt.figure(figsize=(width,height))

```python
plt.figure(figsize=(12,8))

sns.scatterplot(data=df, x='age', y='fare', hue='sex')
```

## Changing the color of the plot

```python
plt.figure(figsize=(16,7))

palette = {'male':'blue', 'female':'red'}

sns.scatterplot(data=df, x='age', y='fare', hue='sex', palette=palette)

plt.savefig('Plots_images/scratter_plot2.png', dpi=300)
```

* When saving a plot save it in 300 dpi or 600 dpi
* Save the plot in png format
#### We can also save the plot in svg format which can be used in webpages and it is vector based and can be zoomed without loss of quality


```python
# Changing the style of the theme in seaborn

sns.set_theme(style='darkgrid', font_scale=1.5)  # Changing font size

sns.scatterplot(data=df, x='age', y='fare', hue='sex')

plt.title('Fare distribution')
```
